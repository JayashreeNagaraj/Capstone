```
import nltk
import re

clean_test =pd.DataFrame(columns = ['id','description'])
def process_clean(text):
  text = text.lower()
  #removing emails
  text = re.sub(r"\S*@\S*\s?"," ", text)
  #removing digits
  text = re.sub(r"\d", " ", text)
  #Detect lang
  #
  #result = translator.translate(text, dest='en')
  #
  #remove all special characters
  text = re.sub(r'[^A-Za-z0-9]+', ' ', text)
  #removing single character
  text = re.sub(r"\s+[a-zA-Z]\s+", " ", text)
  #remove caller from text
  #text = re.sub(caller,' ',text)
  #remove multiple spaces
  text = re.sub(r"\s+"," ", text, flags = re.I)
  #remove space from start and end
  text = re.sub(r"^\s+", "", text)
  # Remove new line characters 
  text = re.sub(r'\n',' ',text)
  text = text.strip()
  return text
clean_test["description"] = x_test1["description"].apply(process_clean)
clean_test["id"] = x_test1["id"]
```

```
import nltk; nltk.download('wordnet'); nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet as wn
from nltk.corpus import stopwords
import string

PUNCT_TO_REMOVE = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~' #string.punctuation
def remove_punctuation(text):
    """function to remove the punctuation"""
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

lemmatizer = WordNetLemmatizer()
def lemma(text):
    pos_dict = {'N': wn.NOUN, 'V': wn.VERB, 'J': wn.ADJ, 'R': wn.ADV}
    return(' '.join([lemmatizer.lemmatize(w,pos_dict.get(t, wn.NOUN)) for w,t in nltk.pos_tag(text.split())]))

#remove punctuation
clean_test["description"] = clean_test["description"].apply(lambda x: remove_punctuation(x))

#Lemmatizer
clean_test["description"] = clean_test["description"].apply(lambda x: lemma(x))
```

```
import tensorflow.keras as tf
from keras.preprocessing.text import Tokenizer, text_to_word_sequence

max_features = 10000

tokenizer = tf.preprocessing.text.Tokenizer( num_words=max_features, filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n', lower=True,
    split=' ', char_level=False, oov_token=None, document_count=0)
texts = tokenizer.fit_on_texts(
     clean_test['description']
 )
sequences = tokenizer.texts_to_sequences(
    clean_test['description']
)
#print(sequences)
desc_seq = tokenizer.get_config()
print(desc_seq)
vocab = tokenizer.word_index
```

```
maxlen=max(len(t) for t in sequences)
print(maxlen)
#maxlen=1500 #Hardcoding length to 1500
data = tf.preprocessing.sequence.pad_sequences(sequences, maxlen=maxlen, padding='post',truncating='post')
```

```
num_words = len(tokenizer.word_index) + 1
embedding_size=100
#print(num_words)

# embeddings_index = {}
# f = open(os.path.join('/content/drive/My Drive/NLP/', 'glove.6B.100d.txt'))
EMBEDDING_FILE = '/content/drive/My Drive/NLP' + '/glove.6B.'+str(embedding_size)+'d.txt'

embeddings = {}
for o in open(EMBEDDING_FILE):
    word = o.split(" ")[0]
    # print(word)
    embd = o.split(" ")[1:]
    embd = np.asarray(embd, dtype='float32')
    # print(embd)
    embeddings[word] = embd

# create a weight matrix for words in training docs
embedding_matrix = np.zeros((num_words, embedding_size))


for word, i in tokenizer.word_index.items():
    embedding_vector = embeddings.get(word)
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector
print(embedding_matrix)
```

```
#from keras.layers import Embedding
from keras.models import Sequential
from keras.layers import Dense, Embedding, LSTM, SpatialDropout1D,Bidirectional
from tensorflow.keras import activations

lstm_out = 196

embedding_layer = Embedding(num_words,
                            embedding_size,
                            weights=[embedding_matrix],
                            input_length=maxlen,
                            trainable=False)

model = Sequential()
model.add(embedding_layer)
model.add(Bidirectional(LSTM(lstm_out, dropout=0.2, recurrent_dropout=0.2)))
model.add(Dense(200,activation=tf.layers.LeakyReLU(alpha=0.2)))
model.add(Dense(100,activation=activations.relu))
#model.add(SpatialDropout1D(0.4))
model.add(Dense(6,activation='softmax'))
```

```
model.compile(loss = 'categorical_crossentropy', optimizer='Adam',metrics = ['accuracy'])
print(model.summary())
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(data, y_test, test_size=0.30)
print(X_train.shape)
print(X_test.shape)
```

```
batch_size = 32
model.fit(X_train, Y_train, epochs = 10, batch_size=batch_size, verbose = 2,validation_split=0.2)
```

```
score,acc = model.evaluate(X_test, Y_test, verbose = 2, batch_size = batch_size)
print("score: %.2f" % (score))
print("acc: %.2f" % (acc))
```
